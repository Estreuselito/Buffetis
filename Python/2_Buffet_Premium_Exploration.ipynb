```python
a = df_returns.iloc[0:12, :]
```

```python
# Create Columns for Event and Dummy 
a.loc[0:, 'Event'] = np.nan
a.loc[0:, 'Dummy'] = np.nan

# Indicate the time of investment
a.loc[6, 'Dummy'] = 1
```

```python
def returns(df):
    length = len(df)
    i = 0
    while i < length:
        if df.loc[i, 'Dummy'] == 1:
            # Add Dummy Variables
            df.loc[i+1,'Dummy'] = 1
            df.loc[i+2,'Dummy'] = 1
            df.loc[i-1,'Dummy'] = 0
            df.loc[i-2,'Dummy'] = 0
            df.loc[i-3,'Dummy'] = 0
            # Add Eventtime
            df.loc[i,'Event'] = 0
            df.loc[i+1,'Event'] = 1
            df.loc[i+2,'Event'] = 2
            df.loc[i-1,'Event'] = -1
            df.loc[i-2,'Event'] = -2
            df.loc[i-3,'Event'] = -3
            i += 3
        else:
            i += 1
        
    df = df.dropna()
    return df
        
```

```python
b = returns(a)
```
